This notebook summarizes the key analytical findings from the supply chain
analysis and provides data-driven recommendations to improve inventory,
shipping, and overall business performance.

## Key Business Insights

### 1. Revenue Concentration Risk
- A small number of products (A-segment) contribute the majority of total net sales.
- XA and YA products are critical revenue drivers.
- Loss or disruption of these products poses significant business risk.

### 2. Demand vs Inventory Mismatch
- Inventory levels remained high even during periods of declining demand.
- Several products show inventory coverage far exceeding actual demand.
- This indicates overstocking and inefficient inventory planning.

### 3. Shipping Performance Issues
- Late shipment rates remained consistently high across shipping modes.
- Fast shipping options (e.g., First Class, Same Day) frequently failed to meet delivery promises.
- Shipping delays are systemic rather than region-specific.

### 4. Volatile Low-Value Products
- Z-segment products exhibit highly unstable demand.
- These products contribute limited revenue but increase operational complexity.


## Data-Driven Recommendations

### 1. Prioritize Core Products (XA, YA)
- Secure alternative suppliers for high-value products.
- Maintain higher service levels and safety stock for these segments.
- Monitor supplier performance closely to prevent future disruptions.

### 2. Optimize Inventory Levels
- Reduce inventory for low-value and volatile segments (Z-products).
- Use inventory coverage ratios to define reorder points.
- Avoid overstocking products with weak or irregular demand.

### 3. Improve Shipping Reliability
- Review and redesign shipping processes for fast delivery modes.
- Align promised delivery times with actual logistics capabilities.
- Consider regional warehouses or third-party logistics partners.

### 4. Implement Demand Forecasting
- Use historical order data to forecast demand at product-segment level.
- Apply different forecasting strategies for X, Y, and Z demand types.
